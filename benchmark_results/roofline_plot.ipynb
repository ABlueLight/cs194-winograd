```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["Naive", "Serial Winograd", "OpenMP Winograd"]
x = [.167, .057, .156]
y = [1.755, .6033, 1.643]
```

```python
fig = plt.figure()
plt.plot([0,3.17], [0, 33.36], color='k', linestyle='-', linewidth=1)
plt.plot([3.17, 10], [33.36, 33.36], color='k', linestyle='-', linewidth=1)
plt.plot(x,y, 'ro')
plt.annotate(
    labels[2], 
    xy = (x[2], y[2]), xytext = (95, -25),
    textcoords = 'offset points', ha = 'right', va = 'bottom',
    bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
    arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
plt.annotate(
    "Peak GFlops (33.36 GFlops)", 
    xy = (3.17, 33.36), xytext = (30, 10),
    textcoords = 'offset points', ha = 'right', va = 'bottom',
    bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
    arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
for label, xs, ys in zip(labels[:2], x[:2], y[:2]):
    plt.annotate(
        label, 
        xy = (xs, ys), xytext = (0, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
ax = fig.get_axes()[0]
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
plt.title("Roofline Model -- Bandwidth 10.53 GB/s")
plt.xlabel("flop/B")
plt.ylabel("GFlop/s")
plt.show()
```
